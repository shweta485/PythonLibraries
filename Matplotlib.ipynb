# Scatter Diagram

```python
age = [5,7,8,7,2,17,2,9,4,11,12,9,6]
speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]
```

```python
plt.scatter(age, speed, marker ='s', color = 'LawnGreen', lw = 4)
plt.show()
```

```python
age =[5,7,8,7,2,17,2,9,4,11,12,9,6]
speed =[99,86,87,88,111,86,103,87,94,78,77,85,86]
speed1 = [85,99,111,86,97,94,87,78,11,55,99,77,15]
```

```python
plt.scatter(age, speed, marker = 'h', c = 'Magenta')
plt.scatter(age, speed1, marker = 's', c = 'CornflowerBlue')
plt.show()
```

# pie diagram

```python
y = [40,10,20,30]
x = ['A','B','C','D']
c = ['LawnGreen','DarkCyan','Crimson','CornflowerBlue']
plt.pie(y,labels = x,colors = c, radius = 2)
plt.show()
```

```python
y = [40,10,20,30]
x = ['A', 'B', 'C', 'D']
c = ['LawnGreen', 'DarkCyan', 'Crimson', 'CornflowerBlue']
plt.pie(y, labels = x, colors = c, radius = 2, explode = [0.0,0.3,0.0,0.0])
plt.show()
```

```python
y = [40,10,20,30]
x = ['A','B','C','D']
c = ['LawnGreen','DarkCyan','Crimson','CornflowerBlue']
plt.pie(y, labels = x , colors = c , radius = 2 , explode = [0.0,0.3,0.0,0.0], startangle = 45)
plt.show()
```

```python
y = [40,10,20,30]
x = ['A','B','C','D']
c = ['LawnGreen','DarkCyan','Crimson','CornflowerBlue']
plt.pie(y, labels = x, colors = c, radius = 2, explode = [0.0,0.3,0.0,0.0], shadow = True)
plt.show()
```

```python
age =[5,7,8,7,2,17,2,9,4,11,12,9,6]
speed =[99,86,87,88,111,86,103,87,94,78,77,85,86]
speed1 = [85,99,111,86,97,94,87,78,11,55,99,77,15]
```

```python
plt.scatter(age, speed, marker = 'h', c = 'Magenta')
plt.scatter(age, speed1, marker ='s', c = 'CornflowerBlue')
plt.xlabel("Age", fontsize = 15)
plt.ylabel("Speed", fontsize = 20)
plt.title("Car comparision", fontsize = 25)
plt.legend(['Speed','Speed1'], loc = "lower right", ncol = 1, facecolor = "yellow", fontsize = 15)
plt.show()
```
